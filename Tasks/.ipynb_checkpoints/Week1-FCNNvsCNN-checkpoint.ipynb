# FCNN versus CNN
This notebook is primarily concerned with recreating Figure 1 in **Why Are Convolutional Networks More Sample Efficient Than Fully-Connected Nets?** by Zhiyuan Li, Yi Zhang and Sanjeev Arora [1]. We have two lines of thought about this figure.

1. We wish to verify that the graphs are authentic and clarify their methods.
2. We believe that the graphs don't reflect the sample-complexity gap claimed in the abstract.

This notebook will address point 1.

## 1. Methods

The original paper somewhat underspecifies their methods, so we mark in **bold** our assumptions and intepretations.

### Data

The inputs are $3\times8\times8$ RGB images. We will use training sets of increasing sizes $10^2,10^3,\dots,10^6$ and **test sets one-fifth the size of the training sets**. The training sets are constructed in the following manner.

+ Entry-wise independent Gaussian (mean 0, standard deviation 1)

The labels are binary, indicating whether the Red channel has larger $l_2$-norm than the Green channel. Mathematically, the label of an image $x$ is $\mathbb{1}[\sum_{i\in R}x_i^2 > \sum_{i\in G}x_i^2]$. As such, the Blue channel has no effect on the true label. 

### Models
1. 2-layer CNN. **IMPORTANT Comment** - *The CNN architecture as described in Figure 1 does not learn at all - I believe they neglected to mention a final fully connected layer.*
    + Input: Three input channels of size 8x8
    + **Convolution: One kernel per input channel of size 3x3, 10 output channels, stride size 1, and padding of 1, and bias**
    + Activation function
    + **Pooling: Max pooling, kernel size 2x2, stride 2**
    + **Flattening**
    + **Fully connected layer (160 in, 1 out) with bias**
    + **Sigmoid**  
2. 2-layer FCNN 
    + Fully connected layer (192 in, 192 out) **with bias**
    + Activation function 
    + Fully connected layer (192 in, 1 out) **with bias**
    + **Sigmoid**
    
For both models, we try both ReLU and Quadratic activation functions (in the convolutional, and hidden layer respectively). 
    
### Training algorithm
+ Stochastic Gradient Descent **with batch size 64**
+ **BCELoss**
+ **Learning rate** $\gamma = 0.01$
+ **Stopping criterion:** 20 epochs.

### Model Evaluation
+ Test loss is measured as MSE over test set.
+ The model $M$ prediction is $\mathbb{1}[M(x)>0.5]$. Test accuracy the percentage of correct predictions over the test set.

```python
# Imports
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader
import random
import numpy as np
import json
```

```python
# Seed random number generation
torch.manual_seed(0)
np.random.seed(0)
random.seed(0)
```

```python
# Global constants
channels = 3 # RGB images
img_size = 8 # Image size length
learning_rate = 0.01
batch_size = 64
epochs = 20
test_fraction = 1/5

# Data size
d = img_size * img_size * 3
N_tr = 100
N_te = int(N_tr * test_fraction)
```

```python
# Data Generation - Normal distribution and Red l2-norm versus Green l2-norm

# Data 
gauss_x_tr = torch.tensor(np.random.normal(0,1,size=(N_tr,channels,img_size,img_size)),dtype=torch.float32)
gauss_x_te = torch.tensor(np.random.normal(0,1,size=(N_te,channels,img_size,img_size)),dtype=torch.float32)

# Labels
square_gauss_tr = torch.sum(gauss_x_tr**2, dim=(2,3))
square_gauss_te = torch.sum(gauss_x_te**2, dim=(2,3))
gauss_y_tr = (square_gauss_tr.T[0] > square_gauss_tr.T[1]).to(torch.float32)
gauss_y_tr = gauss_y_tr.view(gauss_y_tr.size(0),1)
gauss_y_te = (square_gauss_te.T[0] > square_gauss_te.T[1]).to(torch.float32)
gauss_y_te = gauss_y_te.view(gauss_y_te.size(0),1)

# Create a DataLoader for dataset
dataset = TensorDataset(gauss_x_tr, gauss_y_tr)
dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=True)
```

```python
# Define CNN
class CNN(nn.Module):
    def __init__(self, activation):
        super(CNN, self).__init__()
        
        self.conv = nn.Conv2d(3, 10, kernel_size=3, stride=1, padding=1)
        self.activation = activation
        self.pool = nn.MaxPool2d(kernel_size=2, stride=2)
        self.fc = nn.Linear(160, 1)
        self.sigmoid = nn.Sigmoid()
        
        
    def forward(self, x):
        x = self.conv(x)
        x = self.activation(x)
        x = self.pool(x)
        x = x.view(x.size(0), -1)
        x = self.fc(x)
        x = self.sigmoid(x)
        return x
```

```python
# Define FCNN
class FCNN(nn.Module):
    def __init__(self, activation):
        super(FCNN, self).__init__()
        
        self.flatten = nn.Flatten()
        self.fc1 = nn.Linear(d,d)
        self.activation = activation
        self.fc2 = nn.Linear(d, 1)
        self.sigmoid = nn.Sigmoid()
        
        
    def forward(self, x):
        x = self.flatten(x)
        x = self.fc1(x)
        x = self.activation(x)
        x = self.fc2(x)
        x = self.sigmoid(x)
        return x
```

```python
class Quadratic(nn.Module):
    def forward(self, x):
        return x**2
```

```python
# Models
CNNreLU = CNN(nn.ReLU())
CNNquad = CNN(Quadratic())
FCNNreLU = FCNN(nn.ReLU())
FCNNquad = FCNN(Quadratic())
models = [CNNreLU, CNNquad, FCNNreLU, FCNNquad]
names = ["2-CNN+ReLU", "2-CNN+Quadratic","2-FCNN+ReLU","2-FCNN+Quadratic"]

# Optimizers
optimizers = [optim.SGD(model.parameters(), lr=learning_rate) for model in models]
criterion = nn.BCELoss()
```

```python
# Training loop
for i in range(len(models)):
    model = models[i]
    optimizer = optimizers[i]
    model.train()
    for epoch in range(epochs):
        for batch_x, batch_y in dataloader:
            optimizer.zero_grad()
            output = model(batch_x)
            loss = criterion(output, batch_y)
            loss.backward()
            optimizer.step()

        if epoch % 5 == 0:
            print(f"Epoch [{epoch + 1}/{epochs}] Loss: {loss.item()}")
```

```python
# Evaluate models
test_loss = [0 for i in range(len(models))]
accuracy = [0 for i in range(len(models))]

for i in range(len(models)):
    models[i].eval()
    out = models[i](gauss_x_te)
    test_loss[i] += criterion(out, gauss_y_te)
    accuracy[i] += float(sum(torch.eq((out>0.5).to(float),gauss_y_te))/N_te)
```

```python
# Read the JSON file
file_path = 'test_acc.json'
with open(file_path, 'r') as json_file:
    test_acc = json.load(json_file)
    
# Add experiment to results
test_acc[str(N_tr)]=accuracy

# Write accuracy to file
with open(file_path, 'w') as json_file:
    json.dump(test_acc, json_file)
```

### Experiment notes
+ Since I am running on my private laptop I have reduced the synthetic data to 3x8x8.

### Results

```python
import matplotlib.pyplot as plt

# Load data
file_path = 'test_acc.json'
with open(file_path, 'r') as json_file:
    test_acc = json.load(json_file)

# Create sorted dictionary out of saved data
new_dict = {}
for key in test_acc.keys():
    new_dict[int(key)]=test_acc[key]
sorted_dict = dict(sorted(new_dict.items()))

# Convert the dictionary to a format suitable for plotting
x_values = np.sort([int(key) for key in sorted_dict.keys()])  # Convert keys to integers

# For every model, make line plot
for i in range(len(models)):
    
    # The accuracy values for increasing number of samples
    y_values = [value[i] for value in sorted_dict.values()]

    # Create a line plot
    plt.semilogx(x_values, y_values, marker='.', linestyle='-', label=names[i])
    
# Plot graphics
plt.xlabel('Training set size')
plt.ylabel('Test accuracy')
plt.title('Entry-wise normal images')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
